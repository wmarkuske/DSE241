```python
import pandas as pd
```

```python
df = pd.read_csv('./data/co2-population.csv')
df.drop(df[df['State']=='USA'].index, inplace=True)
df.shape
```

```python
states = df.State.unique()
print(states)
sectors = df.Sector.unique()
print(sectors)
years = [1990, 2010]
print(years)
```

```python
headers = ['State']
for year in years:
    for sector in sectors[0:5]:
        header = str(year)+'-'+sector
        headers.append(header)
    headers.append(str(year)+'-Total')
    
for sector in sectors[0:5]:
    headers.append('Diff-'+sector)
headers.append('Diff-Total')

data = pd.DataFrame(columns = headers)
data
```

```python
for state in states:
    row = [state]
    for year in years:
        population = float(df.loc[(df['State']==state) & (df['Sector']=='Population') & (df['Year']==year)].Value.values[0].replace(',',''))
        total = 0
        for sector in sectors[0:5]:
            value = float(df.loc[(df['State']==state) & (df['Sector']==sector) & (df['Year']==year)].Value.values[0])
            total += value
            normal = value/population*100000
            row.append(normal)
        normal = total/population*100000
        row.append(normal)
    for i in range(1,7):
        row.append((row[i+6]-row[i])/row[i]*100)
    data.loc[len(data)] = row

#print(row)
#print(len(row))
data.to_csv('./data/normalized-co2.csv', index=False)
```
