```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import json
#from networkx.readwrte import json_graph
import re
```

```python
G=nx.read_graphml('./data/sheep_ml.graphml')
data = nx.readwrite.json_graph.node_link_data(G)

for node in data['nodes']:
    str=node['id']
    node['id'] = [int(s) for s in str.split("n") if s.isdigit()][0]
    
with open('./data/sheep_ml.json', 'w') as f:
    json.dump(data, f, indent=4)
```

```python
agedf = pd.read_csv('./data/sheep_age.csv')
edgesdf = pd.read_csv('./data/sheep_edges.csv')
```

```python
#G = nx.read_graphml('./data/sheep_ml.graphml')
Graphtype = nx.Graph()
G = nx.from_pandas_edgelist(edgesdf, source='source', target='target', edge_attr='weight', create_using=Graphtype)
nx.draw(G)
```

```python
colors = [i/len(G.nodes) for i in range(len(G.nodes))]
```

```python
nodes = nx.draw(G, cmap=plt.get_cmap('viridis'), node_color=colors)
```

```python
plt.legend(G.nodes())
plt.show()
```

```python
H = nx.DiGraph(G)
nx.draw(H)
```
